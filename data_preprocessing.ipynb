### üî¨ Inspecting transaction.csv

#### üßê Type and Shape

#### üßê Duplication
Next, we check for the duplication of **user_id**

Before we proceed with further analysis, make a copy so that we can manipulate the data without affecting the original.

```python
transaction_df = raw_transaction_df.copy()
```

We combine data and hour column into datetime and convert its type into python datetime64[ns]

```python
transaction_df.insert(0, 'datetime', transaction_df['date'] + ' ' + transaction_df['hour'])
transaction_df = transaction_df.drop(columns=['date', 'hour'])
transaction_df['datetime'] = pd.to_datetime(transaction_df['datetime'], format="%Y-%m-%d %H:%M")
```

```python
transaction_df.info()
```

---

üîé We then check for any **datetime duplications** among the users

```python
print(transaction_df.duplicated(subset=['datetime', 'user_id']).sum())
```

```python
transaction_df.shape
```

There are datetime duplications for some users, and there are two possible cases:
1. The users have the same datetime and the same province (considered a duplicate row, which can be dropped without further analysis).
2. The users have the same datetime but stay in different provinces (Not possible in real life, requires further analysis).

The **first case** yield 465 duplication

```python
print(transaction_df.duplicated(subset=['datetime', 'user_id', 'province']).sum())
```

```python
transaction_df = transaction_df.drop_duplicates(subset=['datetime', 'user_id', 'province'])
```

```python
print(transaction_df.duplicated(subset=['datetime', 'user_id', 'province']).sum())
```

```python
transaction_df.shape
```

After we drop the duplication row is now gone

Now, for the **second** case

```python
users_id = raw_user_df['user_id'].to_list()

# Itterate though all user transaction to see the data
for id in users_id:
    user_transaction = transaction_df[transaction_df['user_id'] == id]
    duplicate_df = user_transaction[user_transaction.duplicated(subset=['datetime'], keep=False)]

    # If this user not have the duplicate row continue to next user
    if len(duplicate_df) == 0:
        continue

    duplicate_df = duplicate_df.sort_values(by=['datetime'], axis=0, ascending=True)
    
    # For extra detail
    hometown = raw_user_df.loc[raw_user_df['user_id'] == id, 'hometown'].iloc[0]
    
    print(f'user id: {id}, hometown: {hometown}')
    display(duplicate_df)
    print()
```

The data show that user exist in difference province but at the same time and some of province is user's hometown.

My suggestion is to **keep the data as it is**, but instead of just sorting by datetime, we should also consider sorting by index. As transactions have no other column to indicate which data come first, **sorting by datetime first and then by the transaction's index** can help establish a clear sequence.

**Noted** that we cannot drop the rows that include user's hometown either because the hometown is necessary to determine the start and end of the user's trip. If we were to drop them, the province_list may become invalid.

```python
transaction_df.shape
```

```python
# We introduce new column name index as key for transaction
transaction_df.insert(0, 'index', range(len(transaction_df)))
```

```python
transaction_df.head(10)
```

```python
# Sorting with user_id then datetime and then index
cleaned_transaction_df = transaction_df.sort_values(by=['user_id', 'datetime', 'index'], axis=0, ascending=True)
```

```python
# We can drop the index column since we done sorting
cleaned_transaction_df = cleaned_transaction_df.drop(columns=['index'])
```

```python
cleaned_transaction_df = cleaned_transaction_df.reset_index(drop=True)
```

```python
cleaned_transaction_df.head(10)
```
